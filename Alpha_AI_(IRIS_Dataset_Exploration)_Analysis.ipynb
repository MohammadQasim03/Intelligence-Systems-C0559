# **Importing The libraries**

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

# **Loading the IRIS Dataset **
 Iris dataset, which is used for multi-class classification tasks where the objective is to predict the species of iris flowers based on their measurements.

```python
# Load the iris dataset
iris = datasets.load_iris()# Loading the Iris dataset from the datasets module
iris_df = pd.DataFrame(iris.data, columns = iris. feature_names)
iris_df. head() # # Displaying the first 5 rows of the DataFrame to inspect the data
```

```python
# Extracting the features from the DataFrame
x = iris_df.iloc[:, [0, 1, 2, 3]].values  # Selecting all rows and the first 4 columns as features

# Importing necessary libraries
from sklearn.cluster import KMeans

# Initializing a list to store the within-cluster sum of squares (WCSS)
wcss = []

# Looping through different values of k to find the optimal number of clusters
for i in range(1, 11):
    # Initializing KMeans with different values of k
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    # Fitting KMeans to the feature data
    kmeans.fit(x)
    # Appending the WCSS value to the list
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt  # Importing the matplotlib library for plotting

# Assuming you already calculated WCSS, let's assume it's stored in a list named 'wcss'

# Plotting the number of clusters (x-axis) against the within-cluster sum of squares (y-axis)
plt.plot(range(1, 11), wcss)  # Corrected 'wcss' from 'WCSS'

# Adding a title to the plot
plt.title('The elbow method')

# Labeling the x-axis
plt.xlabel('Number of clusters')

# Labeling the y-axis
plt.ylabel('WCSS')  # Within-cluster sum of squares

# Displaying the plot
plt.show()
```

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(x)  # Fitting the KMeans model and obtaining the cluster labels
```

```python
# Visualising the clusters - On the first two columns

# Scatter plot for the data points of the cluster labeled as 0 (Iris-setosa)
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=100, c='red', label='Iris-setosa')

# Scatter plot for the data points of the cluster labeled as 1 (Iris-versicolour)
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=100, c='blue', label='Iris-versicolour')

# Scatter plot for the data points of the cluster labeled as 2 (Iris-virginica)
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=100, c='green', label='Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')

# Adding a legend to the plot
plt.legend()

# Displaying the plot
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load the dataset
iris = load_iris()

# Step 2: No preprocessing needed for the Iris dataset

# Step 3: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Step 4: Choose a classification algorithm (Decision Trees)
classifier = DecisionTreeClassifier(random_state=42)

# Step 5: Train the model
classifier.fit(X_train, y_train)

# Step 6: Make predictions
y_pred = classifier.predict(X_test)

# Step 7: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initializing the KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=3)  # You can adjust the number of neighbors as needed

# Training the classifier on the training data
knn_classifier.fit(X_train, y_train)

# Making predictions on the testing data
y_pred_knn = knn_classifier.predict(X_test)

# Evaluating the accuracy of the model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("KNN Classifier Accuracy:", accuracy_knn)
```
